```python
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.title('Netflix Closing Price Trend')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```

```python
# Volatility
df['Daily Returns'] = df['Adj Close'].pct_change()

plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Daily Returns'], label='Daily Returns', color='red')
plt.title('Netflix Daily Returns (Volatility)')
plt.xlabel('Date')
plt.ylabel('Daily Returns')
plt.legend()
plt.show()
```

Days with high volatility are evident through larger spikes in the daily returns

showing  periods of increased risk and opportunity for traders
