# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency = [15, 32, 12, 7]
frequency1 = [30,5,8,50]
fig1, ax1 = plt.subplots(nrows= 1,ncols=2)
explode = (0, 0, 0, 0.2)  # exercise: modify this to "explode" only the 3rd slice (i.e. 'Data Scientists')

# pie 1
ax1[0].pie(frequency, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=90)
ax1[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



#pie #2
ax1[1].pie(frequency1, explode=explode, labels=labels, autopct='%1.1f%%',shadow=True, startangle=180)
ax1[1].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

fig1.tight_layout()
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

# Create a bar plot of performance for each object
plt.xticks(y_pos,objects)
plt.bar(y_pos,performance)

# Use ylabel and title to label the y axis and provide a title
plt.title('Most programming use')
plt.xlabel('Language ')
plt.ylabel('Performance')

plt.show()
```
