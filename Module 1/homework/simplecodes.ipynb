```python
# MEan ,Median and Mode

# Mean = average value
# median = mid point value 
# mode = most common value
from scipy import stats
speed = (99,86,87,85,88,111,87,94,78,77,86,87)

c = np.mean(speed)
d = stats.mode(speed)
e = np.median(speed)
f = np.std(speed)

print(c,"mean",d,"Mode",e,"median",f, "std")
```
